```python
import dateutil
import pandas as pd
```

```python
df_phone = pd.read_csv('phone_data.csv')
df_phone
```

```python
df_phone.groupby('month')['duration'].sum()
```

```python
df_phone[df_phone['item'] == 'call'].groupby('network')['duration'].sum()
#이런식으로 df_phone['item'] == 'call'할게 아니라
#df_phone.groupby(['item','network'])['duration'].sum() #으로 해도 item별로 모인다.
```

```python
df_phone.groupby(['month','item'])['date'].count()
#month -> item으로 묶고 'date' 열에 대해서 count한다.
```

```python
df_phone.groupby(['month','item'])['date'].count().unstack()
#매트릭스 형태로 보자.
```

```python
df_phone.groupby('month', as_index = False).agg({'duration' : 'sum'})
#month로 grouping하고, duration columne에 대해 sum함수를 apply하라.
#as_index = False 이므로 month를 index화 하지 않는다. True라면, month가 index가 된다.
```

```python
df_phone.groupby(['month','item']).agg({'duration' : sum ,
                                        'date' : 'first' ,
                                        'network_type':'count'})
#'month'와 'item'순으로 grouping하고, 각 group에 대해 
#duration의 합계, date의 첫번째 날, network_type의 개수를 출력한다.
```

```python
df_phone.groupby(['month','item']).agg({'network_type':'count',
                                       'date':['min','first','nunique'],
                                       'duration': ['min','max','sum']})
#'month','item'으로 grouping하고, 각 컬럼에 대해 복수 개의 함수를 apply한다. 
#당연히 복수 개의 함수를 전달할 때는 List 형태로 넘긴다.
```

```python
import numpy as np
grouped = df_phone.groupby('month').agg({'duration':['min','max',np.mean]})
#'month'로 grouping하고 'duration' 에 대해 min,max,mean 함수 apply하고
grouped.columns = grouped.columns.droplevel(level=0)
#컬럼 명을 제거해주기 위해 컬럼에서 0번 레벨(맨 첫번째)을 droplevel 해준다.
grouped.rename(columns = {'min' : 'min_duration','max' : 'max_duration','mean':'mean_duration'})
#컬럼명을 제거해준 후, 남은 컬럼명들의 이름을 rename해준다.
```

```python
grouped2 = df_phone.groupby('month').agg({'duration':['min','max',np.mean]})
grouped2.columns = grouped2.columns.droplevel(level=0)
grouped2 = grouped2.add_prefix("duration_")
grouped2
#위에 방법처럼 컬럼에 대해 하나하나 해주지 않고 add_prefix("string")을 해주면 한방에 된다.
```
