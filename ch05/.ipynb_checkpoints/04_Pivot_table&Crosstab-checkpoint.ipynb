```python
import pandas as pd
import dateutil
```

<h1>Pivot table</h1>

```python
df_phone = pd.read_csv('phone_data.csv')
df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst = True)
df_phone.head()
```

```python
df_phone.pivot_table(['duration'],
                    index = ['month', 'item'],
                    columns = 'network', aggfunc='sum').fillna(0)
#index는 인덱스 명을 써도 되고, df_phone.month 처럼 변수로 넘겨도 된다. 
#NaN처리는 매개변수로 fill_value = 0을 주어도 되고, 이전처럼 fillna(0)으로 해도 된다.
```

<h1>CrossTab

```python
df_movie = pd.read_csv('movie_rating.csv')
df_movie.head()
```

```python
df_movie.pivot_table(['rating'], index = df_movie.critic, columns = df_movie.title,
                    aggfunc = 'sum').fillna(0)
#pivot_table로 하면 rating으로 정리한다고 지정하고, 세로 축을 critic 가로축을 title로 지정
```

```python
pd.crosstab(index = df_movie.critic,columns = df_movie.title, values = df_movie.rating,
                    aggfunc='first').fillna(0)
#crosstab으로 하면 세로축(index)은 critic으로 가로축(columns)은 title로
#값(values)은 rating으로 apply할 함수는 first를 사용하고 NaN값은 .fillna를 사용한다.
```

```python
df_movie.groupby(['critic','title']).agg(sum).unstack().fillna(0)
#이런식으로 groupby.unstack으로도 가능하다.
```
